```python
import numpy as np
```

```python
a = np.array(np.arange(1,6))
b = np.array(np.arange(5,10))
```

```python
a
```

```python
b
```

```python
np.hstack((a,b))    # Horizontal Merge (Append)
```

```python
np.column_stack((a,b))   # Horizontal Merge
```
