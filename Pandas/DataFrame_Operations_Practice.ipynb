## **DataFrame Operations**

```python
import numpy as np
import pandas as pd
```

```python
# Creating DataFrame using dictionary
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
# Creating DataFrame using List
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
df2 = pd.DataFrame(data_list)
columns = ["Name","Age","City","Salary"]
df2 = pd.DataFrame(data_list,columns =columns)
df2
```

```python
df2['Name'] # select 1 column
```

```python
df2[['Name','City']]  #Select multiple column
```

```python
df2['Gender']= ["Male", "Female","Male","Female"]   # Creating New Column
```

```python
df2
```

```python
df2.drop('Age', axis=1)    #Remove Coloumn Temporary 
```

```python
df2
```

```python
df2.drop('Age', axis=1, inplace=True)    #Remove Coloumn Permanently 
```

```python
df2
```

```python
df2.drop(['Salary','Gender'], axis=1, inplace=True)    #Remove Multiple Coloumn Permanently 
```

```python
df2
```
